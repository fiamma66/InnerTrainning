### 費氏數列


**『費氏數列』** 是一個每個數等於前兩個數字相加的數列

費氏數列:0, 1, 1, 2, 3, 5, 8, 13, ...

````{admonition} 想法

簡單的概念就是 

只要前後相加就好

所以假設 `a是第一個數` 和 `b是第二個數` 

那麼每次就是 

1. `b 要變成 a + b` 
2. `a 要變成 b` 

```{caution}
要注意先後順序
```

````


### 終極密碼

終極密碼是一個讓對方猜數字，然後我們會把範圍縮小讓他繼續猜的遊戲

我們要自己設定三個東西

1. 最低
2. 最高
3. 答案

這裡可以用 random 幫我們設定好答案

````{note}

python 裡 有內建的 函式庫 random

所以我們要拿來用 就使用 import

*後面我們再詳細講 import*

```python

# 使用 random 中的 function randint(a, b)
random.randint(0,3)

# ranint 會從傳入的兩個整數中(a, b) 隨機取一個
# 必要條件 ： b > a

```
````

## 解答

### 費氏數列 解答

````{admonition} 解答！
:class: dropdown

```python

# 第一個記憶區: 前兩個數 
lasttwo = 0
# 第二個記憶區: 前一個數 
lastone = 0
times = 0
while times < 10:
# 前兩項直接更新就好 
    if times == 0:
     
        lasttwo = 0
        ans = 0
    elif times == 1:
        lastone = 1
        ans = 1
    # 前兩項以後就是把記憶區的東西拿出來 + 
    # 並且更新記憶區
    else:
        ans = lasttwo + lastone
        lasttwo = lastone
        lastone = ans
    print("第", times + 1, "項:", ans) 
    times = times + 1
    
```
````

### 終極密碼解答

````{admonition} 解答！
:class: dropdown

```python

import random 

low = 1
high = 100
ans = random.randint(low, high) 
# 額外記憶區: 記錄猜幾次
times = 0
while True:
# 還記得字串不能跟數字做運算吧? 所以要先把 low, hight 轉成文字再串連 
    msg = "請輸入 [" + str(low) + "-" + str(high) + "]"
    guess = int(input(msg))
    # 次數 + 1
    times = times + 1 
    if guess == ans:
        print("恭喜! 猜對了!")
        break
    elif guess < ans:
    # 猜的比答案小, 更新 low
        low = guess 
        print("狠狠Ｄ 猜錯的次數", times)
    else:
    # 猜的比答案大, 更新 high
        high = guess 
        print("狠狠Ｄ 猜錯的次數", times)
```
````

## 除錯 try catch

各位應該會注意到 

在 python 中 錯了就是直接告訴你錯在哪裡 e.g.

1. ZeroDivisionError  -->  除以 0 的地方
2. ValueError  -->  e.g. 字串無法轉換成 int 時

在剛才的終極密碼中，會發現若我隨意輸入一個 `「 A 」`

會導致 valueError 而結束程式

要避免這種情況，可以用 try catch 來包住 input 的部分

- try catch 來防止輸入非數字

```python
a = int(input())

# ↓↓↓

try:
    a = int(input())
    break
except ValueError:
    print('不是耶！好好輸入數字！')
        
```

這次純粹防止使用者輸入問題

正常自己開發程式時也有必須使用 `try catch` 的地方 e.g.
1. 讀檔案 *（後面詳細講）*
2. 讀 DB 時 處理 Connection Error
